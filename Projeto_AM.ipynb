# Carregando o conjunto de dados usando a biblioteca Pandas

```python
import pandas as pd
import numpy as np
```

importando a base de dados através do github

```python
url = 'https://raw.githubusercontent.com/KacioSilva/ProjetoAM/main/Employee.csv'
df = pd.read_csv(url)
```

#Visualização dos Dados

```python
df.head(15)
```

Visualizando a coluna Education

```python
import matplotlib.pyplot as plt
%matplotlib inline

colunaEducacao = df['Education'].value_counts()
cores = ['c', 'r', 'g']
labels = ['Bacharelado', 'Mestrado', 'PHD']

plt.pie(colunaEducacao, colors=cores, labels=labels, autopct='%1.1f%%', shadow=True)
plt.title('Nível de Formação')
```

Visualizando a coluna Experience in Current Domain

```python
colunaExperiencia = df['ExperienceInCurrentDomain'].value_counts()
cores = ['r', 'g', 'b', 'y', 'c', 'm']

plt.bar(range(len(colunaExperiencia)), colunaExperiencia, width=0.6, color=cores)
plt.title('Distribuição de Anos de Experiência')
plt.ylabel('Quant.Pessoas')
plt.xlabel('Anos de Experiência')
```

Visualizando a distribuição de idades

```python
colunaIdade = df['Age'].value_counts()
idade = colunaIdade.index

plt.bar(idade, colunaIdade, width=0.7, color='b')
plt.title('Distribuição de Idade')
plt.ylabel('Quant.Pessoas')
plt.xlabel('Idade')

plt.xticks(idade)
```

Visualizando a distribuição de sexo

```python
colunaSexo = df['Gender'].value_counts()
cores = ['c', 'm']
labels = ['Masculino', 'Feminino']

plt.pie(colunaSexo, colors=cores, labels=labels, autopct='%1.1f%%', shadow=True)
plt.title('Distribuição por Sexo')
```

Visualizando a distribuição de níveis de Salário

```python
colunaNivelSalario = df['PaymentTier'].value_counts()
salario = colunaNivelSalario.index
cores = ['g', 'b', 'r']

plt.bar(salario, colunaNivelSalario, width=0.4, color=cores)
plt.title('Distribuição dos Níveis de Salário')
plt.ylabel('Quant.Pessoas')
plt.xlabel('Nível')

plt.xticks(salario)
```

```python
colunaSemTrabalho = df['EverBenched'].value_counts()
cores = ['r', 'g']
labels = ['Sim', 'Não']


plt.pie(colunaSemTrabalho, colors=cores, labels=labels, autopct='%1.1f%%', shadow=True)
plt.title("Temporariamente sem Trabalho")
```
